```python
df = pd.read_csv('Data.csv')
```

```python
print(df.head(10))
print(df.isnull().sum())
```

```python
df_part1=df.iloc[1:202,1:26].reset_index(drop=True)
print(df_part1)
df_part2=df.iloc[204:405,1:26].reset_index(drop=True)
print(df_part2)
print(df_part1.equals(df_part2))
# Result is True, meaning there are repeated sections in the original data.
```

```python
df2=df.iloc[204:,1:26].reset_index(drop=True)
df2.columns=['ID','LIMIT_BAL','SEX','EDUCATION','MARRIAGE','AGE','PAY_1','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6','BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6','PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6','DEFAULT']
print(df2)
```

```python
df2['SEX'].replace(['male', 'female'], [1, 2], inplace=True)
df2['EDUCATION'].replace(['graduate school','university','high school','other'],[1,2,3,4],inplace=True)
df2['DEFAULT'].replace(['not default','default'],[0,1],inplace=True)
print(df2)
```

```python
#sex=df2['SEX'].astype('int')
#print(df2['MARRIAGE'].astype('int'))
df3 = df2.astype('int')
print(df3)
```
